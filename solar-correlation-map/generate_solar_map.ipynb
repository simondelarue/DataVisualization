```python
import pandas as pd
```

```python
# Path to data
path = '../../data'
```

```python
df = pd.read_csv(f'{path}/SpeedDating.csv', encoding='latin1')
df.head()
```

```python
df['income']=df['income'].str.replace(',','').astype(float)
df['mn_sat']=df['mn_sat'].str.replace(',','').astype(float)
```

```python
df_match = df[['iid', 'match', 'gender','age', 'int_corr', 
              'attr_o', 'mn_sat', 'income', 'imprelig',
             'go_out', 'date']].groupby(['iid']).mean()
```

```python
df_match_clean = df_match.dropna(axis=0).reset_index(drop=True)
df_match_clean=df_match_clean.rename(columns={'int_corr': 'Same passion', 'attr_o': 'Attractive Score', 'mn_sat':'Scolar level',
                                              'imprelig': 'Same Religion','go_out': 'Often party',
                                              'date': 'Often flirting'})
```

```python
df_match_clean.to_csv('df_match.csv',index=False)
```
